### 기본적인 line plot 그리기

```python
# numpy array로 시각화 데이터 생성

x = np.linspace(-10, 10, 300)
y = x ** 2
```

#### 그래프에 옵션 추가하기
지금부터는 plt 의 다른 함수를 사용하여 보는 사람의 이해를 도울 수 있는 여러 기능들을 알아보겠습니다.
    
- `figure` : 그래프가 그려지는 캔버스 설정  
- `title` : 그래프 제목  
- `xlabel`, `ylabel` : 축 설명  
- `legend` : 범례이름  
- `xlim`, `ylim` : 축 범위  
- `xticks`, `yticks` : 축 구간 내 필요한 구간만 추리기

```python
# 그래프 크기조정 figsize = (가로 세로 사이즈 튜플 전달)
plt.figure(figsize =(8, 6) )
plt.plot(x, y, '--')
plt.show()
```

```python
# 그래프 타이틀 지정
plt.figure(figsize =(8, 6) )
plt.title('$x^2$')
plt.plot(x, y, '--')
plt.show()
```

```python
# 축 타이틀 지정 (축 설명)
plt.figure(figsize =(8, 6) )
plt.title('$x^2$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.plot(x, y, '--')
plt.show()
```

```python
# 범례지정
# 축 범위 설정
plt.figure(figsize =(10, 10) )
plt.title('$x^2$ and $x^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.plot(x, y, '--', label = '$x^2$')
plt.plot(x, x ** 3, '--', label = '$x^3$')
plt.xlim([-5, 5])
plt.ylim([-30, 30])
plt.legend(loc = 'lower right')
plt.show()
```

#### y축에 x의 세제곱 그래프를 그려봅니다. 라인은 노랑색, 선은 실선, 마커는 없어요. 제목은 '$X^3$', plot label=''$x^3$' 범례지정도 해주세요

```python
plt.figure(figsize =(10, 10) )
plt.title('$X^3$')
plt.plot(x, x ** 3, 'y', label = '$x^3$')
plt.legend()
plt.show()
```

#### subplot을 사용하여 여러 그래프 한번에 그리기

```python
x = np.linspace(-10, 10, 300) 



plt.figure(figsize=(12, 8))

plt.subplot(221) # 도화지를 나누는 함수 괄호안에 (도화지의 행, 도화지 열, 순서) 
plt.plot(x, x*x, label='$x^2$')
plt.legend()
plt.ylabel('$x^2$')
plt.xlabel('x')

plt.subplot(222)
plt.plot(x, x*x*x, 'r--', label='$x^3$')

plt.subplot(223)
plt.plot(x, np.sin(x), 'g:', label='$sin(x)$')

plt.subplot(224)
plt.plot(x, np.cos(x), 'k:', label='$cos(x)$')

plt.tight_layout()
plt.show()
```

### bar plot
> 전통적으로 bar plot은 범주형 데이터(구간이 나누어져 있는) 데이터의 갯수, 빈도를 나타내는 데 쓰였습니다.  
지난 시간 학습한 loan 데이터의 신용등급을 bar plot으로 그려보겠습니다.

```python
# loan1.csv 파일 불러들이기
df = pd.read_csv('./data/loan1.csv')
```

```python
# 세부신용등급 각 항목별 갯수세기
df['sub_grade'].value_counts().sort_index()
```

```python
# 세부신용등급 bar plot
plt.figure(figsize=(10, 7))
plt.bar(df['sub_grade'].value_counts().sort_index().index, 
        df['sub_grade'].value_counts().sort_index().values)
```

```python
# 세부신용등급 barh plot
plt.figure(figsize=(10, 7))
plt.barh(df['sub_grade'].value_counts().sort_index(ascending = False).index, 
        df['sub_grade'].value_counts().sort_index(ascending = False).values)
```

### 히스토그램
> 데이터의 분포를 확인하고자 할 때 많이 사용하는 시각화 방법입니다.  
전체 데이터 구간 중 특정 구간에 속한 데이터의 빈도수를 계산하여 구간에 매칭시킵니다.    
데이터 구간별 밀도를 확인할 수 있습니다.

```python
# 대출금액분포 hist plot으로 확인
plt.figure(figsize=(12, 6))
plt.hist(df['loan_amnt'], bins = 50)
plt.show()
```

### scatter plot
> 2개 혹은 2개 이상의 변수간의 관계를 확인하고자 할 때 많이 사용하는 시각화 방법.  
X축과 Y축에 각각의 변수 데이터를 위치시키고 포인트들이 자리한 평면상의 분포를 통해 상관관계를 파악할 수 있습니다.

```python
# 각각의 변수와 우리가 보고자 하는 값과의 관계를 볼 때 가장 많이 사용합니다.
# loan_amnt, installment

plt.figure(figsize=(12, 6))
plt.scatter(df['loan_amnt'], df['installment'])
plt.show()
```

## **seaborn**
> 많은 데이터분석가들이 주로 사용하는 시각화 라이브러리  
matplotlib에 비해 기본적인 시각적 심미성이 뛰어납니다.  
matplotlib에서 지원하지 않는 다양한 형태의 시각화도구를 제공.

### seaborn 불러오기

```python
import seaborn as sns
```

### barplot

```python
plt.figure(figsize=(12, 6))
sns.barplot(df['sub_grade'].value_counts().sort_index().index, 
        df['sub_grade'].value_counts().sort_index().values)
plt.show()
```

### countplot

```python
# 쉬운 방법으로 barplot 구현이 가능하다.
plt.figure(figsize=(12, 6))
sns.countplot(df['grade'],
             order = df['grade'].value_counts().sort_index().index)    # 인덱스 별로 정렬
plt.show()
```

### scatterplot

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(df['loan_amnt'], df['installment'], hue = df["grade"] ,hue_order = ['A', 'B', 'C', 'D', 'E', 'F', 'G'])
plt.show()
```
